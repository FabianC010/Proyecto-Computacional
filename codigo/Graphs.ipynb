```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython import display
import prueba1 as pb

#inicializamos figura y plot para animar e asignamos titulos y limites de los ejes
fig = plt.figure()
line, = plt.plot([])
plt.xlabel('Estados')
plt.ylabel('Probabilidad')
plt.xlim(0, pb.psi.size)
plt.ylim(0, 1)
plt.grid(True)
#funcion encargada de conseguir los valores mas nuevos de la animacion
def animate(frames):
    line.set_data(range(pb.psi.size), pb.stateQuant2[frames*skip])
    return line,

skip = 100
anim = FuncAnimation(fig, animate, frames=(pb.times.size)//skip, interval=42, blit=True)
video = anim.to_html5_video()
html = display.HTML(anim.to_html5_video())
display.display(html)
plt.close()
```
